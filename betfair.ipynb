```python
## BetFair - going to read the main page and grab all the Football game odds. 

import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import display, HTML    # this is so we can display HTML in Jupyter
import re

s = requests.Session()
soup = bs(s.get('https://www.betfair.com/sport/inplay').content, 'html.parser')


LIs = soup.findAll('li', class_ = ['com-coupon-line'])


def report_game(data):
    p0 = re.compile('data-sport="(.+?)"')
    sport = p0.search(data).group(1)
    print(f'{sport} :',end = '')
    if sport != 'SOCCER': return
    p1 = re.compile('data-competition="(.+?)"')
    competition = p1.search(data).group(1)
    p2 = re.compile('data-event="(.+?)"')
    event = p2.search(data).group(1)
    p3 = re.compile('sel-0.+?display-fraction-price">(.+?)</span>.+?sel-1.+?display-fraction-price">(.+?)</span>.+?sel-2.+?display-fraction-price">(.+?)</span>.+?')
    #print(p3.search(data).group(1,2,3))
    hw, draw, aw = p3.search(data).group(1,2,3)
    p4 = re.compile('<span class="home-team-name" title="(.+?)".+?<span class="away-team-name" title="(.+?)"')
    home, away = p4.search(data).group(1,2)

    print(f'  Competition : {competition} \nevent                         : {event}')
    print(f'{home:20} Home-Win : {hw}')
    print(f'{"draw":20} Draw     : {draw}')
    print(f'{away:20} Away-Win : {aw}')

i = 0
for li in LIs:
    print(f'\n {i:3} : -', end='')
    report_game(str(li).replace('\n',''))
    i += 1
```
