```python
walk = open('C:\\users\\clangton2\\downloads\\City2City.kml','r')
lon, lat, km, labels = [], [], [], []


dist = float(0)
i = 0
label = ''
for line in walk:
    line.strip()
    a, b, c = line.split(',')
    lon.append(a)
    lat.append(b)
    
    if i >= 1:
        dist += geopy.distance.geodesic( (a0, b0), (a, b)).km
        label = str(f'{dist:,.3f}km')
    a0 , b0 = a, b
    km.append(dist)
    labels.append(label)
    i += 1
    
df = pd.DataFrame({'lon' :lon, 'lat' :lat, 'km' :km, 'labels': labels})
```

```python

walked = input('Please enter number of KM already travelled :> ')  # How far have they already walked. KM

df1 = df.query(f'km < {walked}')
df2 = df.query(f'km > {walked}')
df2 = df1[-1:].append(df2)

fig = go.Figure(go.Scattermapbox(
    mode = "markers+lines",
    lon = df1.lon,
    lat = df1.lat,
    ids = df1.labels,
    text = df1.labels,
    name=f'completed {walked} KM',
    marker = {'size': 4, 'color':'red'}))

fig.update_traces(
    line=dict(color='red', width=3),
    )

fig.add_trace(go.Scattermapbox(
    mode = "markers+lines",
    lon = df2.lon,
    lat = df2.lat,
    ids = df2.labels,
    text = df2.labels,
    name='road ahead',
    marker = {'size': 4, 'color':'lightblue'}))

fig.update_layout(
    margin ={'l':0,'t':0,'b':0,'r':0},
    mapbox = {
        'style': "stamen-terrain",
        'center': {'lon': float(df2.lon[:1].values[0]), 'lat': float(df2.lat[:1].values[0])},
        'zoom': 4},
)

fig.update_layout(title = 'The long Walk from Cape Town to Moscow')

cities = open('C:\\users\\clangton2\\downloads\\C2Cwaypoints.kml','r')
city, country, lat, lon, e = [], [], [], [], []
for line in cities:
    a, b, c, d, e = line.split(',')
    city.append(f'{a},{b}')
    lat.append(d)
    lon.append(c)

    
fig.add_trace(go.Scattermapbox(
    lat = [ float(df2.lat[:1].values[0]) ], lon = [ float(df2.lon[:1].values[0]) ],
    mode='markers', name='Team 1', text='Team 1',
    marker=dict(size=20, color='red')
))
    
fig.add_trace(go.Scattermapbox(
    lat=lat, lon=lon,
    mode='markers',
    marker=dict(size=14, color='purple'),
    hovertext = city, text=city, name="cities"
))

fig.show()


fig.write_html('C:\\users\clangton2\downloads\City2City.html')
```
