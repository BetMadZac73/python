```python
import pandas as pd
import os
import datetime as dt
from pathlib import Path
folder = f'{Path.home()}\\Downloads\\'
from IPython.display import Markdown, display, HTML
import win32com.client as win32 

df = pd.read_parquet(f'{folder}chg.parquet')
df['CHG START DATE'] = pd.to_datetime(df['CHG START DATE'])
```

```python
df['CHG START DATE'] = pd.to_datetime(df['CHG START DATE'])
```

```python
chg_list = df[((df['CHG START DATE'] > dt.datetime.now() - dt.timedelta(days=3)) & 
             (df['CHG START DATE'] < dt.datetime.now() + dt.timedelta(days=3)))]

# i is for the index position in the location we are cutting out
def cut_location(x, i):
    if type(x) == str:
        octets = x.split('/')
        if len(octets) > i: return octets[i]
    return 'Unknown'

chg_list['COUNTRY'] = chg_list.apply(lambda x: cut_location(x['CHG LOCATION'], 3), axis=1)
chg_list['SITE'] = chg_list.apply(lambda x: cut_location(x['CHG LOCATION'], 5), axis=1)



display(HTML(chg_list.to_html()))
```
